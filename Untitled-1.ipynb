```python
# import all necesary libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import folium
```

```python
print('The Dataset Contains, Rows: {:,d} & Columns: {}'.format(listings.shape[0], listings.shape[1]))
```

```python
#Checking for null values in columnes of interest 

columns = ['review_scores_rating']

nulls = listings[columns].isnull().sum()
print(nulls)
```

```python
listings['review_scores_rating'].info()

print('porcentaje de nulos en los datos:', 4362/18320*100)
```

```python
import matplotlib.pyplot as plt

# Contar los valores nulos en la columna 'review_scores_rating'
null_counts = listings['review_scores_rating'].isnull().sum()

# Contar los valores no nulos en la columna 'review_scores_rating'
non_null_counts = listings['review_scores_rating'].notnull().sum()

# Configurar los datos para el gráfico de barras
labels = ['Non-null values', 'Null values']
values = [non_null_counts, null_counts]

# Trazar el gráfico de barras
plt.bar(labels, values, color=['blue', 'red'])

# Añadir título y etiquetas
plt.title('Null Values in review_scores_rating Column')
plt.ylabel('Count')

# Mostrar el gráfico
plt.show()


print('nulls val:', null_counts,)
print('non nulls val:', non_null_counts,)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un histograma para la columna review_scores_rating
sns.histplot(data=listings, x='review_scores_rating', kde=True)

# Añadir etiquetas y título al gráfico
plt.xlabel('Review Scores Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Review Scores Rating')

# Mostrar el gráfico
plt.show()
```
